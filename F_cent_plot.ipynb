```python
import math
import matplotlib.pyplot as plt
import numpy as np

m_beam = 108.35 #g
m_beam = m_beam/1000 #kg
L = 0.29 #m
lambd = m_beam / L

rpm = 300
omega = 2 * math.pi * rpm / 60

def F_at_x(x):
    return lambd * x * omega**2 *(L - x/2)

# plot F(x) for x in [0, L]
x = np.linspace(0, L, 100)
F = [F_at_x(xi) for xi in x]
plt.plot(x, F)
plt.xlabel('x (m)')
plt.ylabel('F(x) (N)')
plt.title('Normal force on beam section as a function of x')
plt.show()


```
